```python
# --- Configuración Inicial ---
saldo = 1000  # Saldo inicial ficticio
Cajero = []  # La lista empieza vacía [1]
continuar_en_cajero = True

print("-----------------------------------------")
print("¡Bienvenido al Cajero Inteligente AI-2026!")
print("-----------------------------------------")

# Bucle principal interactivo [2, 3]
while continuar_en_cajero:
    print("\n--- MENÚ DE OPCIONES ---")
    print("1. Consultar Saldo")
    print("2. Depositar Dinero")
    print("3. Retirar Dinero")
    print("4. Ver Historial (Lista)")
    print("5. Salir")

    opcion = input("\nSelecciona una opción (1-5): ")

    # Estructuras de decisión para el menú [4, 5]
    if opcion == "1":
        print(f"\n[INFO] Tu saldo actual es: ${saldo}")

    elif opcion == "2":
        deposito = input("¿Cuánto deseas depositar?: ")
        monto_deposito = float(deposito)
        if monto_deposito > 0:
            saldo = saldo + monto_deposito
            # Agregamos la operación a la lista [1]
            Cajero_memoria.append(f"Depósito: +${monto_deposito}")
            print("¡Depósito exitoso!")
        else:
            print("[ERROR] El monto debe ser mayor a 0.")

    elif opcion == "3":
        retiro = input("¿Cuánto deseas retirar?: ")
        monto_retiro = float(retiro)
        # Verificación lógica de fondos suficientes [5, 6]
        if monto_retiro <= saldo:
            if monto_retiro > 0:
                saldo = saldo - monto_retiro
                Cajero_memoria.append(f"Retiro: -${monto_retiro}")
                print("Retirando dinero... Por favor, toma tu efectivo.")
            else:
                print("[ERROR] Monto inválido.")
        else:
            print("[ERROR] Fondos insuficientes.")

    elif opcion == "4":
        print("\n--- Tu Historial de Transacciones ---")
        # Usamos while y un contador porque el ciclo 'for' está prohibido [3, 7]
        indice = 0
        largo_lista = len(Cajero_memoria)

        if largo_lista == 0:
            print("La lista está vacía. Aún no hay movimientos.")
        else:
            while indice < largo_lista:
                print(f"- {Cajero_memoria[indice]}")
                indice = indice + 1

    elif opcion == "5":
        print("Gracias por usar el Cajero Inteligente. ¡Hasta pronto!")
        continuar_en_cajero = False # Rompe el ciclo while [3]

    else:
        print("[ERROR] Opción no válida, intenta de nuevo.")

print("\n--- Sesión Finalizada ---")
```
