## Realizando simulacion de inteligencia artificial con if, else, elif  while tipo Menu
# Problema numero 15: Cajero

```python
# --- SIMULACIÓN DE CAJERO AUTOMÁTICO INTELIGENTE (IA) ---

# Configuración inicial de datos (Simulando la base de datos)
nombre_cliente = "Alex"
saldo_cuenta = 5000.0
pin_secreto = 1234

# REQUISITO: La lista empieza completamente vacía.
# Servirá como la "Memoria de Movimientos" para que la IA analice al usuario.
historial_movimientos = []

print("IA: ¡Hola! Soy tu Asistente Financiero con Inteligencia Artificial.")
print("IA: Detecto tu presencia frente al cajero. Por favor, identifícate.")

# Entrada interactiva para el PIN
pin_ingresado = int(input("CLIENTE: Ingresa tu PIN de 4 dígitos: "))

# Verificación de identidad
if pin_ingresado == pin_secreto:
    print(f"\nIA: ¡Acceso concedido! Analizando perfil... Bienvenido/a, {nombre_cliente}.")

    # Control del ciclo (1 = Activo, 0 = Salir)
    estado_cajero = 1

    # El ciclo while continuará mientras estado_cajero sea igual a 1
    while estado_cajero == 1:
        print("\n==========================================")
        print(f"IA: Tu saldo actual es: ${saldo_cuenta}")
        print("IA: ¿Qué acción deseas realizar hoy?")
        print("Opciones: 1. Retirar | 2. Depositar | 3. Consultar Memoria de IA | 4. Salir")
        print("==========================================")

        opcion = input("CLIENTE (Elige una opción del 1 al 4): ")

        # Opcion 1: Retirar Dinero
        if opcion == "1":
            print(f"\nIA: Has elegido Retirar. Tu saldo disponible es de ${saldo_cuenta}.")
            monto_retirar = float(input("CLIENTE: ¿Cuánto dinero deseas retirar?: $"))

            if monto_retirar > 0:
                if saldo_cuenta >= monto_retirar:
                    saldo_cuenta = saldo_cuenta - monto_retirar
                    print(f"IA: ¡Retiro exitoso! Toma tu dinero.")

                    # Guardamos la acción en la lista (Memoria de la IA)
                    historial_movimientos.append(f"Retiro de ${monto_retirar}")
                else:
                    print("IA: Alerta. No tienes fondos suficientes para esa cantidad.")
            else:
                print("IA: Error. No puedes retirar cantidades negativas o en cero.")

        # Opcion 2: Depositar Dinero
        elif opcion == "2":
            print("\nIA: Has elegido Depositar.")
            monto_depositar = float(input("CLIENTE: ¿Cuánto dinero vas a ingresar al cajero?: $"))

            if monto_depositar > 0:
                saldo_cuenta = saldo_cuenta + monto_depositar
                print(f"IA: ¡Depósito recibido con éxito! Tu nuevo saldo se ha actualizado.")

                # Guardamos la acción en la lista (Memoria de la IA)
                historial_movimientos.append(f"Deposito de ${monto_depositar}")
            else:
                print("IA: Error. El monto a depositar debe ser mayor a 0.")

        # Opcion 3: Consultar Memoria de la IA (Uso de la lista)
        elif opcion == "3":
            print("\nIA: Accediendo a mi registro de Inteligencia Artificial...")

            # Verificamos si la lista sigue vacía usando un if
            if historial_movimientos == []:
                print("IA: Mi memoria está vacía. Aún no has realizado movimientos en esta sesión.")
                print("IA: Consejo IA: Realiza un depósito o retiro para poder analizar tu comportamiento.")
            else:
                print(f"IA: He analizado tu sesión. Tus movimientos registrados son: {historial_movimientos}")

                # Un toque de "IA" que reacciona según el tamaño de la lista
                cantidad_movimientos = len(historial_movimientos)
                print(f"IA: Diagnóstico: Has realizado {cantidad_movimientos} operaciones.")

                if cantidad_movimientos >= 3:
                    print("IA: Consejo IA: Veo que estás muy activo hoy. ¡Cuida de no gastar de más!")
                else:
                    print("IA: Consejo IA: Mantienes un flujo financiero moderado y saludable.")

        # Opcion 4: Salir
        elif opcion == "4":
            print("\nIA: Cerrando sesión y guardando datos de forma segura...")
            estado_cajero = 0

        # Opción no válida
        else:
            print("\nIA: Error de comando. Por favor, elige una opción válida (1 al 4).")

else:
    print("\nIA: Alerta de seguridad. El PIN es incorrecto. Acceso denegado.")

print("\nIA: Gracias por usar el Cajero Inteligente. Procesadores en modo de espera. ¡Adiós!")
```
